# II. Linear Algebra and Probability Theory

#### Linear Algebra

1.	Linear Algebra is a form of continuous mathematics. Scalars, vectors and matrices and tensors are important basic structures used in Machine Learning. A __Hadamard Product__ or element-wise production is a literal matrix as a product of two matrices (a matrix containing product of individual elements), denoted by A ⊙ B. We also revise some properties of matrices like associativity, distributivity, etc.
2.	An identity matrix contains ones throughout its main diagonal and zeroes in the remaining slots. Multiplying this matrix to any vector doesn’t change the value of the resulting vector <br>
<center>A x = B <br>
A<sup>-1</sup> A x = b A<sup>-1</sup> <br>
I<sub>n</sub> x = b A<sup>-1</sup>	&ensp;&ensp;	[A A<sup>-1</sup> = I<sub>n</sub>] <br>
x = A<sup>-1</sup> b <br></center>
3.	Despite useful properties of A<sup>-1</sup>, its use is discouraged in software applications because A-<sup>-1</sup> can be represented with only limited precision on computers.
4.	To analyse, how many solutions exist of equation A x = B, think of columns of matrix A as specifying directions we can travel in from origin, then determine how many ways there are of reaching b <br>
<center>A x = ∑ I  x<sub>i</sub> A<sub>:,i</sub> <br>
This is a linear combination. <br>
The span of a set of vectors is the set of all points obtainable by linear combination of original vectors.</center>
5.	 Having n >= m and matrix columns being linearly independent are necessary and sufficient conditions to find solutions. A square matrix with linearly dependent columns is known as singular.
6.	In Machine Learning, we use a function called as __Norm__, to measure the size of vectors. The norm of a vector x measures the distance from the origin to point x.
7.	The L<sup>2</sup>  norm is Euclidean Norm, denoted by ||x|| which is simply the Euclidean distance from origin to the point identified by x. Whenever the difference between zero and non-zero elements is important, the simple L<sup>1</sup> norm is used
<center>||x||<sub>1</sub> = ∑<sub>i</sub> |x<sub>i</sub>|</center>
8.	The L<sup>∞</sup> Norm, known as max norm, simplifies to absolute value of element with largest magnitude in vector. Size of a matrix is measured using Frobenius norm, which involves finding the square root of summation of square of each element in the matrix.
9.	A diagonal matrix is any matrix with non-cero values along its diagonal and xero values in remaining places. A symmetric matrix is any matrix that is equal to its own transpose.
10.	If product of transpose of a vector with another vector is zero, then they are orthogonal vectors. An orthogonal matrix is a square matrix whose inverse is equal to its transpose.
11.	Eigen decomposition is a kind of matrix decomposition, in which we decompose a matrix into a set of eigenvectors an eigenvalues.
12.	An eigenvector of a square matrix A is a non-zero vector v such that multiplication by matrix A alters only the scale of v. Decomposing a matrix with its eigenvalues helps us analyse certain hidden properties of the matrix. Every real symmetric matrix can be decomposed into only real valued eigenvectors and eigenvalues
<center>A = Q ^ Q <sup>T</sup><br>
where Q is orthogonal matrix and ^ is a diagonal matrix. </center>
13.	Single Value Decomposition (SVD) provides a way to factorize a matrix into single vectors and single values. If eigen decomposition is not possible, we can use SVD instead. It is more generalized.
14.	The trace operator gives the sum of all diagonal entries of a matrix. This is very useful to find ways around arriving to certain equations with other missing data.
15.	Principle Component Analysis is a simple machine leaning algorithm derived using basic linear algebra rules as stated above. In this our aim is to apply lossy compression to an array of points by introducing minimal deviation from precision and hence making it suitable to be stored in low memory.


#### Probability Theory

1.	Machine learning makes use of probability theory concepts heavily, because it must frequently deal with stochastic and uncertain data. Uncertainty can result from many factors like incomplete modelling, incomplete observability of environment or even inherently stochasticity in the system.
2.	A probability distribution over discrete variables is a probability mass function. Probability mass functions can act on many variables at the same time. This PMF   over many variables is known as a joint probability distribution.
3.	At the same time, a probability distribution over continuous variables is a probability density function. The probability distribution over the subset of variables is known as the marginal probability distribution. Conditional probability is probability of certain event occurring assuming that another dependent event has already occurred.
4.	The chain rule states that, any joint probability distribution over many random variables may be decomposed into conditional distributions over only one variable. 
5.	The expectation of a function with respect to a probability distribution is the average that the function takes on when the event occurs. The variance gives a measure of how much the values of a function of a random variable vary as we sample diﬀerent values of x from its probability distribution. The covariance gives some sense of how much two values are linearly related to each other.
6.	Many probability distributions are extensively used in machine learning. The Bernoulli distribution is a distribution over a single binary random variable. It is controlled by a single parameter, which gives the probability of random variable being equal to 1. 
7. The Multinoulli distribution is a distribution over a single discrete variable with diﬀerent finite states. The exponential distribution helps us obtain probability distribution with a sharp point at x= 0. A closely related probability distribution that allows us to place a sharp peak of probability mass at an arbitrary point is the Laplace distribution.
8.	Dirac delta function is useful for specifying mass in a probability distribution cluster s around a single point. Sometimes probability distributions are combined to get a  better understanding, this is known as mixture distributions. 
9. On each trial, the choice of which component distribution should generate the sample is determined by sampling a component identity from a multinoulli distribution.
10.	The logistic sigmoid is commonly used to produce the φ parameter of a Bernoulli distribution. It saturates when its argument is very positive or very negative, meaning that the function becomes very ﬂat and insensitive to small changes in its input.

### Preprocessing of data
For the sake of visualizing all data, we are converting the non-numeric data to a numeric form for better visual rendering.

The Formatted date column contains date in the form of YYYY-MM-DD HH:MM:SS +Z. Such precision is unecessary for a vast database visualization. Hence, hereafter only the year and month of each entry will be accounted for. The method __year_month_conv__ is used for manipulating the string and converting it to YYYYMM numeric format.

```python
def year_month_conv(s):
    s = s[:7]
    if '-' in s:
        s = s[:4] + s[5:7]
    return s               

regdf['Formatted Date'] = regdf['Formatted Date'].apply(lambda x: year_month_conv(x))
```

__create_dict__ is a generic method for converting every unique string value in a column, into a dictionary entry against a number. For example: {'Partly Cloudy': 1, 'Mostly Cloudy':2} <br>
This method will be used throughout the notebook.

```python
def create_dict(unique_set):
    new_dic = {}
    for i in range(unique_set.size):
        new_dic.update({unique_set[i]:i})         
    return new_dic
```

```python
summary_unique = regdf['Summary'].unique()
precip_unique = regdf['Precip Type'].unique()
daily_sum_unique = regdf['Daily Summary'].unique()
```

```python
summary_dic = create_dict(summary_unique)
precip_dic = create_dict(precip_unique)
daily_sum_unique = create_dict(daily_sum_unique)
```

```python
print(summary_dic)
print(precip_dic)
print(daily_sum_unique)
```

Prerpocessing the entries of three textual columns using the create dictionary method for __Summary__, __Precipe Type__ and __Daily Summary__. 

```python
X = regdf.iloc[:, :12].copy()
X['Summary'] = X['Summary'].apply(lambda x: summary_dic[x])
X['Precip Type'] = X['Precip Type'].apply(lambda x: precip_dic[x])
X['Daily Summary'] = X['Daily Summary'].apply(lambda x: daily_sum_unique[x])
X.head()
```

```python
df = pd.read_csv("AtchisonUV_20150801_to_20151119.csv")
df.shape
```

```python
df.head()
```

### Preprocessing of data
For the sake of visualizing all data, we are converting the non-numeric data to a numeric form for better visual rendering.

The __Date__ column contains date in the form of YYYY-/MM-DD HH:MM:SS (AM/PM). Such precision is unecessary for a vast database visualization. Hence, hereafter only the month of each entry will be accounted for, since all entries are from the same year. The method __year_conv__ is used for manipulating the string and converting it to MM numeric format.

```python
def year_conv(s):
    s = s[:2]
    if '/' in s:
        s = str(0) + s[:1]
    return s                
X = df.iloc[:, :10].copy()
X['Date'] = X['Date'].apply(lambda x: year_conv(x))
X.head()
```

The __Wind Origin__ column alone contains textual data, hence we are using the __create_dict__ method to obtain its word to number dictionary and use it below. 


```python
stabilitydic = create_dict(X['Wind Origin'].unique())
print(stabilitydic)
```

```python
X['Wind Origin'] = X['Wind Origin'].apply(lambda x: stabilitydic[x])
X.head()
```

### Visualization

```python
fig = plt.figure(figsize=(16,12))
plt.clf()
for i in range(10):
    plt.subplot(4, 3, i+1)
    plt.plot(X.iloc[:, i])
    plt.ylabel(X.columns.values[i])
fig.tight_layout()
```

### Observation
1. Benzene levels fluctuate from 2 to 10 in the first hundredd entries. After which it's consistent around 2.5
2. CS2 levels are a constant at 2.5 throughout. We can ignore this attribute in future anaysis, because of lack of fluctuation to impact the pattern at all
3. Except for first few entries, Ozone content mostly falls between 0 to 150
4. SO2 level slowly rise upto 300 value
5. Toluene and Xylene exhibit similar levels after a first few entries
6. Wind Direction has no pattern and is spanned out in all directions erratically
7. Wind Speed falls between 1 to 18 in most cases
8. Wind Origin looks as erratic as the wind direction graph
